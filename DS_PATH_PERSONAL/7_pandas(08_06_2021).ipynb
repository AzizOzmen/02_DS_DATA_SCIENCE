```python
def multiply(n):
    return n *2
```

```python
df["tip*3"]=df["tip"].apply(lambda x: x*3)
```

```python
def quality(total_bill,tip):
    if tip/total_bill  > 0.25:
        return "Generous"
    else:
        return "Other"
```

```python
df["t_quality"]=np.vectorize(quality)(df["total_bill"],df["tip"])
df[df["t_quality"]=="Generous"]
```
